```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
from keras import Sequential
model = Sequential()
print(model)
```

```python
from keras import layers
```

```python
model.add(layers.Dense(units=4, input_shape=(10,)))
model.summary()
```

```python
#Implementacje funkcjii aktywacji

from keras.activations import linear
random_data = sorted(np.random.randn(200))
df = pd.DataFrame({"data":random_data, "linear":linear(random_data)})
df.head()
```

```python
px.line(df)
```

```python
px.line(df, x="data", y="linear", width=800, range_y=[-2,2])
```
